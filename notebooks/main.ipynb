```python
import pandas as pd

# Load the CSV file (adjust the path because you're in the notebooks directory)
df = pd.read_csv('../data/assets_data_mapped.csv')

# Select only the important columns
df_clean = df[['id', 'name', 'asset_type_name', 'asset_tag', 'department_name', 'created_at', 'end_of_life', 'user_id', 'description']]

# Drop any duplicates
df_clean = df_clean.drop_duplicates()

# Convert Dates to a Consistent Format 
df_clean['created_at'] = pd.to_datetime(df_clean['created_at'], errors='coerce')
df_clean['end_of_life'] = pd.to_datetime(df_clean['end_of_life'], errors='coerce')

# Identify text columns (i.e., object type columns) to standardize text
text_columns = df_clean.select_dtypes(include=['object']).columns

# Standardize text columns: strip whitespace, convert to lowercase (or title case if preferred)
df_clean[text_columns] = df_clean[text_columns].apply(lambda x: x.str.strip())  # Remove leading/trailing spaces
df_clean[text_columns] = df_clean[text_columns].apply(lambda x: x.str.lower())  # Convert to lowercase

# Preview the cleaned dataset
df_clean.head()

# Save the cleaned data back to the data directory
df_clean.to_csv('../data/cleaned_asset_data.csv', index=False)
```

```python
import pandas as pd

# Load datasets
df_filewave = pd.read_csv('../data/cleaned_filewave_data.csv')
df_headcount = pd.read_csv('../data/headcount_data.csv')

# Create 'full_name' in df_headcount
df_headcount['full_name'] = (df_headcount['First Name'] + ' ' + df_headcount['Last Name']).str.strip().str.lower()

# Ensure 'full_name' in df_filewave is cleaned
df_filewave['full_name'] = df_filewave['Last Logged Username'].fillna('').astype(str).str.strip().str.lower()

# Clean 'full_name' in df_headcount
df_headcount['full_name'] = df_headcount['full_name'].fillna('').astype(str).str.strip().str.lower()
```

```python
from thefuzz import fuzz, process

def fuzzy_merge(df_left, df_right, left_on, right_on, threshold=80, limit=1):
    # Drop NaN values and convert to strings
    s_left = df_left[left_on].dropna().astype(str).drop_duplicates().reset_index(drop=True)
    s_right = df_right[right_on].dropna().astype(str).drop_duplicates().reset_index(drop=True)
    
    matches = []
    for left_value in s_left:
        # Skip non-string and empty values
        if not isinstance(left_value, str) or not left_value.strip():
            continue
        match = process.extractOne(left_value, s_right, scorer=fuzz.token_sort_ratio)
        if match and match[1] >= threshold:
            matches.append({'left_value': left_value, 'right_value': match[0], 'score': match[1]})
    return pd.DataFrame(matches)
```

```python
# Normalize names
df_filewave['normalized_name'] = df_filewave['full_name'].str.replace(r'\W+', '', regex=True).str.lower()
df_headcount['normalized_name'] = df_headcount['full_name'].str.replace(r'\W+', '', regex=True).str.lower()

# Fuzzy merge using partial_ratio
matches = fuzzy_merge(df_filewave, df_headcount, 'normalized_name', 'normalized_name', threshold=50)

# Merge the matched records back to the original DataFrames
df_filewave_matched = pd.merge(df_filewave, matches, left_on='full_name', right_on='left_value', how='left')

df_filewave_headcount = pd.merge(
    df_filewave_matched,
    df_headcount,
    left_on='right_value',
    right_on='full_name',
    how='outer',
    suffixes=('_filewave', '_headcount')
)

# Save matches DataFrame
matches.to_csv('../data/matches_filewave_headcount.csv', index=False)

# Save merged DataFrame
df_filewave_headcount.to_csv('../data/df_filewave_headcount.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of similarity scores
matches['score'].hist(bins=20)
plt.xlabel('Similarity Score')
plt.ylabel('Number of Matches')
plt.title('Distribution of Similarity Scores')
plt.show()
```
