```python
import pandas as pd

# Load the CSV file from the data directory
data_path = '../data/assets_data_mapped.csv'  # Adjust the path relative to the notebook directory
data = pd.read_csv(data_path)

# Check the first few rows of the dataset
data.head()
```

```python
# Basic info about the data
data.info()

# Check for missing values
data.isnull().sum()

# Summary statistics for numeric columns
data.describe()
```

```python
# Unique values in department_name
print("Unique departments:", data['department_name'].nunique())
print(data['department_name'].value_counts())

# Unique asset names
print("Unique asset names:", data['name'].nunique())
print(data['name'].value_counts())  # Top 10 most frequent asset names
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot for department distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='department_name', data=data, order=data['department_name'].value_counts().index)
plt.title('Distribution of Assets by Department')
plt.xlabel('Number of Assets')
plt.ylabel('Department')
plt.show()
```

```python
# Group by department and asset type to see the most common asset types per department
department_asset_group = data.groupby(['department_name', 'asset_type_name']).size().reset_index(name='count')

# Sort and show the top asset types in each department
department_asset_group = department_asset_group.sort_values(by='count', ascending=False)
print(department_asset_group.head(10))

# Visualization of the top asset types by department
plt.figure(figsize=(12, 8))
sns.barplot(x='count', y='department_name', hue='asset_type_name', data=department_asset_group)
plt.title('Top Asset Types by Department')
plt.xlabel('Count')
plt.ylabel('Department')
plt.legend(title='Asset Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
